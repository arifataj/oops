Inheritance allows us to define a class that inherits all the methods and properties from another class . Parent class is the class being inherited from ,also called base class.
child clss is the class that inherits from another class , also called derived class . Reuse of code .Add extra functionality to existing class

parent = base class

```python
class A :
  def printa (self ,num) :
    self.num = num
    print('inside class A')

class B(A) :
  def printb (self):
    print('inside class B')

obj =  B()

obj.printa('hey1234')
obj.printb()
```

**single inheritance**

Single inheritance allows a derivate class to inherit properties of one parent class,and this allows code reuse and the introduction of additional features in existing code .

**Multilevel Inheritance**

The featurs that are part of the original class, as well as the class that is derived from it, are passed on to the new class. It is similar to releshionship

```python
class A :
  def printa(self):
    print('inside class A')
class B(A):
  def printb(self):
   print('inside class B')

class C(B):
  def printc(self) :
    print('inside class C')

obj = C()

obj.printa()
obj.printb()
obj.printc()
```

**Multiple Inheritance**

if a class is able to be created from multiple base classes , this kind of inheritance is known as multiple inheritance . When this is mutiple inheritance ,each of the atribute  that are present in the classes of the base has been past on to the class that is derived from it

```python

class A :
  def printa(self):
    print('inside class A')
class B:
  def printb(self):
   print('inside class B')

class C(A,B):
  def printc(self) :
    print('inside class C')

obj = C()

obj.printa()
obj.printb()
obj.printc()
```

**Super()function**

Python also has a super()function that will make the child class inherit all the methods and properties from its parent

```python
#create a parent class :
class FirstClass:
 def __init__(self,v1,v2):
  print('I am in FirstClass')
  self.v1 =v1
  self.v2 =v2
 def printValue(self):
  print('values are', self.v1, 'and', self.v2)
```

```python
#SecondClass inheriting from class FirstClass

class SecondClass(FirstClass):
  def __init__(self,v1,v2,v3):
    self.v3 =v3
    super().__init__(v1,v2)
    def printValue(self):
      print('I am in SecondClass')
    def printValue(self):
      print('I am in SecondClass printValue')
```

**ENCAPSULATION**

Encapsulation binds together data and functions that maniputate the data,keeping it safe from interference and misuese.

Encapsulation is one of the fundamental concepts in object-oriented programming (OOP).
It describes the idea of wrapping data and the motheds that works on data within one unit.This puts restrictions on accessing varibles and methods directly and can prevent the accidental modification of data. To prevent accidental change ,an objects variable can only be changed  by an objects mothod. Those types of variables are known as private variables

self.name - Public

self._name-Protected.-Private

self._name = privat---strongly private

```python
class Encapsulation():
  def __init__(self,a,b,c):
    self.a = a
    self._b = b
    self.__c = c

  def mymethod(self):
    print(self.__c)
```

* **ABSTRACTION IN PYTHON**

Abstraction is used to hide the internal functionality from the users. The users only intract with the basics implementation of function ,but inner working is hidden .Users is familier with that **"what function does"** but they don't know **"how it does"**

```python
from abc import ABC,abstractmethod

class Mail(ABC):

  @abstractmethod
  def mileage(self):
    pass

  def fun(self):
    print('fun')

class Child(Mail):

  def sum (self):
    print('sum')

  def mileage(self):
    print('mileage')

obj = Child()
obj.mileage()
```
